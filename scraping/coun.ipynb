```python
url = "https://abbreviations.yourdictionary.com/articles/country-abbreviations.html"
```

```python
html = requests.get(url, headers= headers1)
soup = BeautifulSoup(html.text, "lxml")
```

```python
con = []
for tr in soup.find_all('tr'):
    tds = tr.find_all('td')
    con.append(tds[2].text.lower())
```

```python

textfile = open("country.txt", "w",encoding='utf-8')
for element in con:
    textfile.write(element + "\n")
textfile.close()
```
